```python
%matplotlib inline
# %matplotlib inline sets the backend of matplotlib to the 'inline' backend
import matplotlib.pyplot as plt
# Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy

# Seaborn style on matplotlib plot, menentukan jenis graph. Untuk jenis graph lain bisa dilihat di
# https://python-graph-gallery.com/199-matplotlib-style-sheets/
import numpy as np
from sklearn.cluster import KMeans
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca dataset student_lifestyle.csv
dataset = pd.read_csv('student_lifestyle.csv')

# Pilih dua fitur untuk divisualisasikan
# Ganti nama kolom sesuai dengan dataset Anda
features = dataset[['Study_Hours_Per_Day', 'Sleep_Hours_Per_Day']]

# Membuat scatter plot dari dua fitur yang dipilih
plt.scatter(features['Study_Hours_Per_Day'], features['Sleep_Hours_Per_Day'])

# Berikan label sumbu x dan y
plt.xlabel('Study Hours Per Day')
plt.ylabel('Sleep Hours Per Day')

# Berikan judul pada plot
plt.title('Scatter Plot of Student Lifestyle Features')

# Tampilkan plot
plt.show()
```

```python
X = np.array(list(zip(features['Study_Hours_Per_Day'], features['Sleep_Hours_Per_Day']))).reshape(len(features['Sleep_Hours_Per_Day']), 2)
SSE = []
K = range(1, 10)
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(X)
    kmeanModel.fit(X)
    SSE.append(kmeanModel.inertia_)
```

```python
# Plot the elbow
plt.plot(K, SSE, 'bx-')
plt.xlabel('k')
plt.ylabel('SSE')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

```python
y_pred = KMeans(n_clusters=4).fit_predict(X)
plt.plot()
LABEL_COLOR_MAP = {
    0: 'r',
    1: 'g',
    2: 'b',
    3: 'y'  # Add a color for cluster 3
}

label_color = [LABEL_COLOR_MAP[l] for l in y_pred]
plt.scatter(features['Study_Hours_Per_Day'], features['Sleep_Hours_Per_Day'], c=label_color)  # Use column names for features
plt.title("k=4")
plt.show()
```

```python
figure, ax = plt.subplots(2, 3)
K = range(1, 7)
for k in K:
    if k < 4:
        row = 0
        column = k - 1
    else:
        row = 1
        column = k - 4
    kmeanModel = KMeans(n_clusters=k).fit(X)
    y_pred = kmeanModel.fit_predict(X)
    # Access features using column names instead of numeric indices
    ax[row][column].scatter(features['Study_Hours_Per_Day'], features['Sleep_Hours_Per_Day'], c=y_pred, cmap='winter_r')

plt.show()
```
