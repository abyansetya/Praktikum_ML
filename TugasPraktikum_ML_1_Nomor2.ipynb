```python
from sklearn.model_selection import train_test_split # Import the train_test_split function
```

```python
# Assuming 'dataset' is a pandas DataFrame
array = dataset.values

X = array[:,0:4]
Y = array[:,4]

# Convert Y to integers or categorical if it's not already
Y = pd.Series(Y).astype('category').cat.codes.values  # Convert to category codes

validation_size = 0.20
seed = 7

# Now you can call train_test_split
X_train, X_validation, Y_train, Y_validation = train_test_split(X, Y, test_size=validation_size, random_state=seed)
```

```python
seed = 7

scoring = 'accuracy'
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

models = []

models.append((
  'KNN',
    KNeighborsClassifier(n_neighbors=5)
))

models.append((
    'NB',
    GaussianNB()
))

models.append((
    'SVM',
    SVC(gamma='auto')
))

results = []

# Use a different variable name for the list of model names
model_names = []

for name, model in models:
  kfold = KFold(n_splits=10, random_state=seed, shuffle=True)
  cv_results = cross_val_score(model, X_train, Y_train, cv=kfold, scoring=scoring)
  results.append(cv_results)
  # Append the model name to the 'model_names' list
  model_names.append(name)

  msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
  print(msg)
```

```python
from sklearn.metrics import accuracy_score # Import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

svm = SVC(gamma='scale')
svm.fit(X_train, Y_train)
predictions = svm.predict(X_validation)
print(accuracy_score(Y_validation, predictions))
print(confusion_matrix(Y_validation, predictions))
print(classification_report(Y_validation, predictions))
```
